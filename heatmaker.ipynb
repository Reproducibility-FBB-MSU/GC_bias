```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("/home/ferryman/Загрузки/project/df_plant.csv", index_col=0)
cleardf=df.drop(["sum","gc_ended_sum"], axis=1)
hmd={key : np.corrcoef([cleardf[key],cleardf["gc_fraction"]])[0,1] for key in cleardf.columns}
hmd=pd.DataFrame(pd.Series(hmd))
hmd=hmd.drop("gc_fraction")
mtx=sorted(list(hmd.index))
mtd=[hmd.loc[elem] for elem in mtx]
mtd=np.array(mtd).reshape(16,4)
mtx=np.array(mtx).reshape(16,4)
u=list(mtd[:,3])
mtd[:,3]=mtd[:,1]
mtd[:,1]=u
ind=[elem[0:2] for elem in mtx[:,0]]
col=["A","U","G","C"]
mtd=pd.DataFrame(data=mtd, index=ind, columns=col)
sns.set()
sns.heatmap(mtd)
```
